### **Atividade** - Como se chama o robô?


```haskell
[2,5,2,3] |> sort
```

```haskell
[2,5,2,3] |> maximum
```

```haskell
[2,5,2,3] |> sum
```

```haskell
[2,5,2,3] |> nub
```

```haskell
[2,5,2,3] |> reverse
```

```haskell
[2,5,2,3] |> length
```

### **Atividade** - Que robôs podem dar a mão?


```haskell
[4,1,1,2] |> nub |> reverse
```

```haskell
[4,1,1,2] |> reverse |> length
```

```haskell
[4,1,1,2] |> maximum |> sum
```

```haskell
[4,1,1,2] |> length |> maximum
```
