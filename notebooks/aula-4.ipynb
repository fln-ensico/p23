## **Introdução**

### Nesta atividade vamos considerar o seguinte:

* ### **um símbolo** entre **`''`** é um **caracter** (e só podemos ter um símbolo entre **`''`**):
    * ### Exemplo 1: **`'a'`**;
    * ### Exemplo 2: **`'2'`**.
* ### **um ou mais símbolos** entre **`""`** é uma **sequência de caracteres** (uma **palavra** ou **frase**):
    * ### Exemplo 1: **`"a"`**;
    * ### Exemplo 2: **`"aluno"`**;
    * ### Exemplo 3: **`"o meu primeiro computador"`**.

## **Exercício 1** | Língua 01

### A **língua 01** consiste em transformar:

* ### a letra minúscula **`'o'`** no caracter (número) **`'0'`**:
    * ### **`'o'`** $\rightarrow$ **`'0'`**
* ### a letra minúscula **`'i'`** no caracter (número) **`'1'`**:
    * ### **`'i'`** $\rightarrow$ **`'1'`**

### **Exemplos:** 
* ### **`"computador"`** $\rightarrow$ **`"c0mputad0r"`**
* ### **`"motivo"`** $\rightarrow$ **`"m0t1v0"`**

### **a)** Vamos começar por criar uma **função** chamada **`muda01`** que permita fazer as transformações acima.

#### **Nota:** Os caracteres diferentes de **`'o'`** e  **`'i'`** não devem ser alterados (devem ficar iguais a eles próprios).

```haskell
muda01 'o' = '0'
muda01 ... = ...
muda01 outro_caracter = ...
```

### **a.2)** **Executa** as células abaixo para confirmares que a função **`muda01`** está bem definida.

### **b)** E se quisermos aplicar as transformações acima à palavra **`"limpo"`**?
### Será que **`muda01 "limpo"`** nos permite fazer isso? 
### Testa essa hipótese **executando** a célula abaixo.

```haskell
muda01 "limpo"
```



### **Ups!** Dá erro... Porquê? 🤔

### A função **`muda01`** recebe um **caracter** e dá como resultado um **caracter**. Não sabe fazer mais nada!



### Mas **`"limpo"`**, como já sabemos, não é um **caracter**, mas sim uma **sequência de caracteres**, certo?

### Vamos investigar o assunto para perceber bem o que isso significa🕵️‍♂️

### **c)** **Executa** a célula abaixo. Qual será o seu significado?

```haskell
:t 'o'
```

### <span style="color:darkblue"> O computador está a dizer-nos que **`'o'`** é uma coisa do tipo **`Char`**. Mas o que é isso? 🤔 </span>

### "**`Char`**" é uma abreviatura da palavra **`Char`acter**, que significa **caracter** em inglês 😉

### **d)** E qual será o tipo de **`"limpo"`**? **Executa** a célula abaixo.

```haskell
:t "limpo"
```

### O computador está a dizer-nos que **`"limpo"`** é uma **sequência de caracteres**.

#### Por isso é que **`Char`** surge delimitado por **`[`** e **`]`**, surgindo **`[Char]`**.



### ⚠️ Então, **`muda01 "limpo"`** dá erro porque:
* ### a função só sabe transformar **caracteres**...
* ### mas **`"limpo"`** é uma **palavra**, ou seja, é uma **sequência de caracteres**.

### <span style="color:darkblue"> Mas que sequência é essa, mesmo? 😅 </span>

* ### **`'e'`** $\rightarrow$ **`'E'`**;
* ### **`'i'`** $\rightarrow$ **`'I'`**;
