### **f)** Assim, para traduzir uma frase em português para a língua do Yoda, temos de:
* ### obter a **sequência de palavras** que a constitui, recorrendo à função **`words`**;
* ### **inverter** essa sequência de palavras, recorrendo à função **`reverse`**;
* ### transformar a sequência obtida novamente numa **frase**, recorrendo à função **`unwords`**.



### **f.1)** Escreve uma expressão que permita traduzir a frase **`"vou comer arroz"`** para a língua do Yoda.

#### **Completa** a célula abaixo.

```haskell
"vou comer arroz" |> ... |> ... |> ...
```

### **f.2)** Vamos dar o nome **`linguaYoda`** à função que acabamos de criar.

#### O símbolo **`>>`** é utilizado para **encadear** funções, i.e., para obter uma função que é o resultado de aplicarmos várias funções sucessivamente.



### **Executa** a célula abaixo.

```haskell
linguaYoda = words >> reverse >> unwords 
```
