## **Exercício** | Língua do P

### A **língua do P** consiste em introduzir, após cada **vogal**, a consoante **`'p'`** seguida por essa **vogal**.

### **Exemplos:** 
* ### **`"escola"`** $\rightarrow$ **`"epescopolapa"`**
* ### **`"computador"`** $\rightarrow$ **`"copompuputapadopor"`**



### No fundo, as **vogais** são transformadas nas **palavras** seguintes:

* ### **`'a'`** $\xrightarrow{muda}$ **`"apa"`**;
* ### **`'e'`** $\xrightarrow{muda}$ **`"epe"`**;
* ### **`'i'`** $\xrightarrow{muda}$ **`"ipi"`**;
* ### **`'o'`** $\xrightarrow{muda}$ **`"opo"`**;
* ### **`'u'`** $\xrightarrow{muda}$ **`"upu"`**.

### **a)** Recorda que uma **palavra** é uma **sequência de letras**.

```haskell
['d']
```

### **g)** Resumindo, o que fizemos foi:
* ### começar com uma **palavra**;
* ### aplicar o "feitiço" (ou função) **`map muda`**, obtendo uma **sequência de palavras**;
* ### aplicar o "feitiço" (ou função) **`concat`** para converter a **sequência de palavras** obtida numa **palavra** só.



### Para dizermos ao computador que queremos aplicar a função **`map muda`** e, de seguida, a função **`concat`**, escrevemos:

* ### **`map muda >> concat`**. 

### **g.1)** Vamos agora dar o nome **`linguaP`** à função que acabaste de criar.

```haskell
linguaP = map muda >> concat
```
