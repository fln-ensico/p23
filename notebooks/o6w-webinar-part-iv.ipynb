## **Part IV** - Dictionaries

---


```haskell
portuguese = ["gato", "casa", "ovo", "raposa"]
```

```haskell
english = ["cat", "house", "egg", "fox"]
```

```haskell
portuguese # english
```

```haskell
english # portuguese
```

```haskell
a = [1,2,3]
b = [4,5,6]
c = [7,8,9]
```

```haskell
a # b
```

```haskell
b # a
```

```haskell
a # (b # c)
```

```haskell
(a # b) # c
```

---


```haskell
dict = [("cat","felino"), ("cat","gato"), ("egg","ovo"),("fox","raposa"), ("house","casa"), ("residence","casa")]
```

```haskell
dict |> collect
```

---


```haskell
elvish = ["aerister", "bhin", "bhin", "lurleath", "nor"]
english = ["teacher", "young man", "boy", "sky", "love"]
```

```haskell
(elvish # english) |> collect
```

```haskell
(elvish # elvish) |> collect
```

```haskell
(elvish # elvish) |> collect |> length
```
