#### ➡️ Começamos por dizer ao computador **quem** são os utilizadores do *Instagram*:

```haskell
data Pessoas = Maria | Rui | Isabel | Joana | Pedro deriving (Eq, Ord, Show)
```

---


---
#### Uma possível **sequência** (a ordem dos elementos podia ser outra):

```haskell
instagram = [(Maria, Isabel),
            (Maria, Rui),
            (Rui, Maria),
            (Rui, Isabel),
            (Rui, Joana),
            (Isabel, Pedro),
            (Joana, Isabel),
            (Joana, Pedro),
            (Joana, Rui),
            (Pedro, Isabel)]
```

---


---
#### ➡️ A sequência abaixo indica quantas pessoas cada um dos jovens **segue**:

```haskell
quantos_segue = [(x, length (filter (== x) instagram')) | x <- nub instagram']
    where instagram' = map fst instagram
quantos_segue
```

---
#### ➡️ A sequência abaixo indica quantos **seguidores** cada um dos jovens tem:

```haskell
num_seguidores = [(x, length (filter (== x) instagram'')) | x <- nub instagram'']
    where instagram'' = map snd instagram
num_seguidores
```

#### ➡️ A **sequência** passa a ser:
