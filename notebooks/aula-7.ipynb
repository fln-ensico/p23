### ➡️ Cifra de César

```haskell
suc 'Z' = 'A'
suc ' ' = ' '
suc c = chr(ord c+1)

unsuc 'A' = 'Z'
unsuc ' ' = ' '
unsuc c = chr(ord c-1)

cesar = suc >> suc >> suc
uncesar = unsuc >> unsuc >> unsuc

cifraCesar = map cesar
decifraCesar = map uncesar
```

```haskell
cifraCesar "JOAO"
```

```haskell
decifraCesar "VRQKD H DPELFLRQD"
```

---
### ➡️ Cifra "A-Z"

```haskell
transfAZ c = chr (f n) where
   n = ord c
   f n = ord 'Z' - (n - ord 'A')
   
cifraAZ = map transfAZ
decifraAZ = cifraAZ
```

```haskell
decifraAZ "VMHRXL"
```
