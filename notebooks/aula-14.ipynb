### ➡️ **Sequências**

```haskell
[80,35,...]
```

* #### Podemos colocar espaços entre as vírgulas e os números, não interfere:

```haskell
[80, 35, 123, 35, 18]
```

```haskell
[80  ,    35,123, 35,  18]
```

* #### Mas o que acontece se nos esquecermos dos parênteses **`[ ... ]`**?

```haskell
80,35,123,35,18
```

* #### O que acontece se colocarmos pontos **`.`** em vez de vírgulas **`,`**?

```haskell
[80.35.123.35.18]
```

#### **Conclusão:** Temos de escrever tudo direitinho, com **parênteses** e **vírgulas**! 😃
