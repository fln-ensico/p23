* ### **`'o'`** $\rightarrow$ **`'u'`**;
* ### **`'u'`** $\rightarrow$ **`'a'`**.

### **b)** E se quisermos aplicar a troca de vogais acima à palavra **`"escola"`**?
#### Será que **`trocas "escola"`** nos permite fazer isso? Testa essa hipótese **executando** a célula abaixo.

```haskell
trocas "escola"
```



### **Ups!** Dá erro... Porquê? 🤔

### ⚠️ O problema é que **`'a'`**, **`'e'`**, **`'i'`**, etc. são **caracteres** (mais precisamente, letras). Mas **`"escola"`** é uma **palavra**, não é um caracter!



### A função **`trocas`** recebe um **caracter** e dá como resultado um **caracter**. Não sabe fazer mais nada!



```haskell
:t trocas
```

```haskell
:t "escola"
```
