### 👉 True power is achieved when we **compose functions** (i.e. successively apply one function after another).

### For instance, if we want to compute the **sum** of all the **distinct** numbers in the list **`[3,2,7,1,7,2,4]`**, we write:

```haskell
[3,2,7,1,7,2,4] |> nub |> sum
```
