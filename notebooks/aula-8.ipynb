#### ➡️ Vamos dizer ao computador quais são as várias **fases da obra**:

```haskell
data Obras = Paredes | Telhado | Soalho | Portas | Colocar_portas | Pintar | Ok
             deriving Show
```

#### ➡️ **Sequência** que representa o grafo da obra:

```haskell
obra =
   [((Portas,Colocar_portas),1),
    ((Paredes,Colocar_portas),5),
    ((Paredes,Telhado),5),
    ((Paredes,Soalho),5),
    ((Colocar_portas,Pintar),1),
    ((Telhado,Pintar),4),
    ((Soalho,Pintar),3),
    ((Pintar,Ok),2)
   ]
```

#### ➡️ Vamos pedir ao computador para **desenhar** este grafo:

```haskell
obra |> desenha_grafo
```
