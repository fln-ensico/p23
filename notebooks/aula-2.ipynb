```haskell
a |-> b = (a,b)
set = sort . nub
agrupa :: (Ord b, Ord a) => [(b, a)] -> [(b, [a])]
agrupa x = set [ k |-> set [ d' | (k',d') <- x , k'==k ] | (k,d) <- x ]
```
