```haskell
data Pessoas = Ana | Helena | Manuel | Teresa | Paulo deriving (Eq, Ord, Show)
data Roupa = Calças | Vestido deriving (Eq, Ord, Show)
```

```haskell
fila = [Ana, Helena, Manuel, Teresa, Paulo]
roupa = [Calças, Calças, Calças, Vestido, Calças]
```

```haskell
fila # roupa
```

```haskell
alturas = map snd [(Ana, 1.6), (Helena, 1.7), (Manuel, 1.8), (Teresa, 1.6), (Paulo, 1.9)]
```

```haskell
alturas
```

```haskell
fila # roupa
```

```haskell
roupa # fila
```

```haskell
(fila # roupa) # alturas
```
